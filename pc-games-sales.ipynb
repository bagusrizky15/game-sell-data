```python
#Importing libraries for Exploratory data analysis and plotting

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Importing Dataset
df = pd.read_csv("../input/most-selling-pc-games/Games.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Checking null values
df.isnull().sum()
```

```python
df.Series.value_counts()
```

```python
#Dropped null values
df.dropna(how="any",inplace=True)
```

```python
df.isna().sum()
```

```python
# Histogram for distribution of sales
df.Sales.hist();
```

```python
#plotting bar graph for top 5 most popular genre
most_popular_genre = pd.DataFrame(df.Genre.value_counts()[:5])
most_popular_genre.plot(kind="bar",figsize=(15,5),color=("lightgreen"))
plt.title("Top 5 Most Popular Genre",fontsize=(20))
plt.xlabel("Genre",fontsize=(20))
plt.ylabel("Count",fontsize=(20))
plt.legend()
plt.xticks(rotation=0);
print("")
```

```python
#plotting bar graph for top 5 games publisher
Top_publisher = pd.DataFrame(df.Publisher.value_counts()[:5])
Top_publisher.plot(kind="bar",figsize=(15,5),color=("salmon"))
plt.title("Top 5 Games Publisher",fontsize=(20))
plt.xlabel("Publisher",fontsize=(20))
plt.ylabel("Count",fontsize=(20))
plt.xticks(rotation=0);
```

```python
#plotting barh graph for top 10 games developer
Top_developer = pd.DataFrame(df.Developer.value_counts()[:10])
Top_developer.plot(kind="barh",figsize=(15,5),color=("lightblue"))
plt.title("Top 10 Games Developer",fontsize=(20))
plt.xlabel("Count",fontsize=(20))
plt.ylabel("Developer",fontsize=(20))
plt.xticks(rotation=0);
```

```python
#plotting bar graph for most popular series
most_popular_series = pd.DataFrame(df.Series.value_counts()[:5])
most_popular_series.plot(kind="bar",figsize=(15,5),color=("grey"))
plt.title("Most Popular Series",fontsize=(20))
plt.xlabel("Series",fontsize=(20))
plt.ylabel("Count",fontsize=(20))
plt.xticks(rotation=0);
```
